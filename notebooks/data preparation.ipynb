## cleaning

```python
cdt.title = cdt.title.str.lower()
```

```python
cdt['country'] = cdt['location'].str.split(', ').str[-1]
```

```python
edu_amount_cdt = edu[['candidate_id', 'education_id']].groupby('candidate_id').nunique().reset_index()
```

```python
edu_amount_cdt.rename(columns={
    'education_id':'degrees_amt'
}, inplace=True)
```

```python
edu_amount_cdt
```

## merging

```python
cdt = pd.merge(cdt, edu[['candidate_id', 'education_id']], on='candidate_id', how='left')
```

```python
cdt['has_degree'] = cdt['education_id']
```

```python
cdt.has_degree = cdt.has_degree.notnull().astype('int')
cdt
```

```python
edu = pd.merge(edu, edu_amount_cdt, on='candidate_id', how='left')
edu
```
